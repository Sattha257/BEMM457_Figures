```python
#Figure 2 Retention Rate by Age Group
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/Users/satth/Desktop/Cleaned_Age_Data.csv'
cleaned_age_data = pd.read_csv(file_path)

# Filter data for retention rates by age group
age_retention = cleaned_age_data[cleaned_age_data['Metric'] == 'Retention Rate']

# Create a figure with a specified size
plt.figure(figsize=(12, 8))

# Iterate over unique age groups and plot their retention rates
for age_group in age_retention['Age_at_Entry'].unique():
    # Subset the data for the current age group
    subset = age_retention[age_retention['Age_at_Entry'] == age_group]
    # Select the years (columns starting from 2013 to 2022)
    years = cleaned_age_data.columns[4:]
    # Get the retention rate values and convert them to float
    rates = subset.iloc[0, 4:].values.astype(float)
    # Plot the retention rates with a line and marker
    plt.plot(years, rates, marker='o', label=age_group)

# Add a title and labels for the axes
plt.title("Retention Rates by Age Group (2013-2022)")
plt.xlabel("Year")
plt.ylabel("Retention Rate")

# Add a legend and move it outside the plot area for better visibility
plt.legend(title="Age Group", bbox_to_anchor=(1.05, 1), loc='upper left')

# Add a grid to the plot for easier readability
plt.grid(True)

# Adjust layout to make space for the legend
plt.tight_layout()

# Display the plot
plt.show()
```

```python
#Figure 11 Forecasting Retention Rates by Age Group 
import pandas as pd
from statsmodels.tsa.holtwinters import ExponentialSmoothing
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/Users/satth/Desktop/Cleaned_Age_Data.csv'
data = pd.read_csv(file_path)

# Filter out rows where 'Age_at_Entry' is missing or labeled as 'Age Missing'
data = data[~data['Age_at_Entry'].str.contains('Unknown|Age Missing', na=False)]

# Define the available years in the dataset
years = ['2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']

# Extract unique age groups
age_groups = data['Age_at_Entry'].unique()

# Create a plot for each age group
plt.figure(figsize=(12, 8))

forecasts = []

for age_group in age_groups:
    # Filter data for each age group
    filtered_data = data[
        (data['Metric'] == 'Retention Rate') &
        (data['Age_at_Entry'] == age_group) &
        (data['Starting_Enrollment_Intensity'] == 'Total')
    ]
    
    # Proceed only if data is available for the age group
    if not filtered_data.empty:
        # Extract year columns and convert to time series
        time_series = filtered_data[years].iloc[0].values
        dates = pd.to_datetime(years)
        
        # Create DataFrame for the current age group
        ts_df = pd.DataFrame({'Date': dates, 'Retention Rate': time_series})
        ts_df.set_index('Date', inplace=True)
        
        # Fit Holt-Winters' Exponential Smoothing model
        hw_model = ExponentialSmoothing(ts_df['Retention Rate'], trend='add').fit()
        hw_forecast = hw_model.forecast(steps=3)
        
        # Plot the original data and forecast
        plt.plot(ts_df.index, ts_df['Retention Rate'], label=f'Observed ({age_group})')
        plt.plot(hw_forecast.index, hw_forecast, linestyle='--', label=f'Forecast ({age_group})')
        
        # Collect forecast data
        forecasts.append({
            'Age Group': age_group,
            'Year': ['2023', '2024', '2025'],
            'Holt-Winters Forecast': hw_forecast.values
        })

# Customize the plot
plt.xlabel('Year')
plt.ylabel('Retention Rate')
plt.title('Retention Rate Forecasting for Age Groups using Holt-Winters Method')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # Position the legend outside the plot
plt.grid(True)
plt.tight_layout()  # Adjust layout for better fit
plt.show()

```
